```python
#2 cols, CG, IG
import Hornet
import time
from datetime import datetime as dt
import csv

now = dt.now()
filename = "pressure_log_"+str(now.strftime('%Y%m%d%H%M'))+".csv"
print("starting pressure log file: "+filename)

with open(filename, 'a') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerow(['# DateTimestamp','CGPressure(Torr)', 'IGPressure(Torr)'])

    while True:
        CGpressure = Hornet.getConvectronP()
        IGpressure = Hornet.getIonP()
        now = dt.now()
        nowstr = now.strftime('%Y%m%dT%H%M%S')
        
        if str(Hornet.IG_stat())[11:-4] != "ON":
            IGpressure = 'NaN'
            
        print(nowstr, CGpressure, IGpressure)
        csvwriter.writerow([nowstr, CGpressure,IGpressure])
        
        # flush the file
        csvfile.flush()

    csvfile.close()
```
